```python
import opensmile
import audiofile
```

```python
files = []
for audio in audios:
        signal, sampling_rate = audiofile.read(audio, always_2d=True)
        files.append([signal, sampling_rate])
```

```python
smile = opensmile.Smile(
    feature_set=opensmile.FeatureSet.eGeMAPSv02,
    feature_level=opensmile.FeatureLevel.Functionals,
)
# smile.feature_names
len(smile.feature_names)
```

```python
categories = ['eGeMAPSv02_Functionals', 'eGeMAPSv02_LowLevelDescriptors' , 'ComParE2016_LowLevelDescriptors']


Audio__eGeMAPSv02_Functionals_paths = []

for f_index, file in enumerate(files):
    data = smile.process_signal(
        file[0],
        file[1]
    )
    for p_index, path in enumerate(audios):
        if (f_index == p_index):
            data_path = path.replace('Bag_of_lies_video.wav', '')
            csv_path = f'{data_path}Audio__eGeMAPSv02_Functionals.csv'
            Audio__eGeMAPSv02_Functionals_paths.append(csv_path)

            data.reset_index(inplace=True)
            data.drop(columns=['start', 'end'], inplace=True)
            
            data.to_csv(csv_path, index=False)
            break
    

annotations=pd.read_csv(f'{base_dir}/Annotations.csv')
annotations['Audio__eGeMAPSv02_Functionals'] = pd.DataFrame(Audio__eGeMAPSv02_Functionals_paths)
annotations.to_csv(f'{base_dir}/Annotations.csv', index=False)
annotations.head()
```

```python
smile = opensmile.Smile(
    feature_set=opensmile.FeatureSet.eGeMAPSv02,  
    feature_level=opensmile.FeatureLevel.LowLevelDescriptors,
)
smile.feature_names
len(smile.feature_names)
```

```python
audio_paths = []

for f_index, file in enumerate(files):
    data = smile.process_signal(
        file[0],
        file[1]
    )
    for p_index, path in enumerate(audios):
        if (f_index == p_index):
            data_path = path.replace('Bag_of_lies_video.wav', '')
            csv_path = f'{data_path}Audio.csv'
            audio_paths.append(csv_path)
            
            columns = [categories[1] + '__' + column for column in data.columns]
            
            data.columns = columns
            data.reset_index(inplace=True)
            data.drop(columns=['start', 'end'], inplace=True)
            
            if (os.path.isfile(csv_path)):
                old_df = pd.read_csv(csv_path)
                old_df[columns] = data
                old_df.to_csv(csv_path, index=False)
                
            else:
                data.to_csv(csv_path, index=False)
            break

annotations=pd.read_csv(f'{base_dir}/Annotations.csv')
annotations['Audio'] = pd.DataFrame(audio_paths)
annotations.to_csv(f'{base_dir}/Annotations.csv', index=False)
```

### Audio Finalization

```python
audio_csv_list=pd.read_csv(f'{base_dir}/Annotations.csv')['Audio'].tolist()

Values = []
columns = []

for filename in audio_csv_list:
    df=pd.read_csv(filename)
    
    if (len(columns)<1):
        columns = df.columns
        
    maximum = df.max().tolist()
    minimum = df.min().tolist()
    mean = df.mean().tolist()
    median = df.median().tolist()
    std = df.std().tolist()
    var = df.var().tolist()
    percentile25 = df.quantile(0.25).tolist()
    percentile50 = df.quantile(0.5).tolist()
    percentile75 = df.quantile(0.75).tolist()
    skew=df.skew().tolist()
    kurto=df.kurtosis().tolist()
    li=maximum+minimum+mean+median+std+var+percentile25+percentile50+percentile75+skew+kurto
    Values.append(li)
```

```python
stats=['max','min','mean','median','std','var','per25','per50','per75','skew','kurt']
cols=[col+'_'+stat for stat in stats for col in columns]

df1=pd.DataFrame(data=Values,columns=cols)
df1.head()
```

```python
audio__eGeMAPSv02_Functionals_list=pd.read_csv(f'{base_dir}/Annotations.csv')['Audio__eGeMAPSv02_Functionals'].tolist()

values = []
columns = []

for path in audio__eGeMAPSv02_Functionals_list:
    df = pd.read_csv(path)
    if (len(columns) < 1):
        columns.append(df.columns.tolist())
    values.append(df.values.tolist()[0])


df2 = pd.DataFrame(data=values, columns=columns[0])
df2.head()
```

```python
columns = ['eGeMAPSv02_Functionals__' + column for column in df2.columns]
df2.columns = columns
df2.head()
```

```python
df = pd.concat([df1, df2], axis=1)
```
